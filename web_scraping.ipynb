```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.hubertiming.com/results/2018MLK"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a', href = True)
print(links)
```

```python
for link in links:
  print(link['href'])
```

```python
data = []
allrows = soup.find_all("tr")
for row in allrows:
  row_list = row.find_all("td")
  dataRow = []
  for cell in row_list:
    dataRow.append(cell.text)
  data.append(dataRow)
titles = data[0]
data = data[4:]
print(titles)
print(data[-2:])
```

```python
df = pd.DataFrame(data)
print(df.head())
print(df.tail())
```

```python
hearder_list = []
col_hearders = soup.find_all('th')
for col in col_hearders:
  hearder_list.append(col.text)
print(hearder_list)
```

```python
df.columns = hearder_list
print(df.head())
```
